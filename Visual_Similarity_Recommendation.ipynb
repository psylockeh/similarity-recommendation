```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from datetime import datetime
import numpy as np

# Configurações
img_width, img_height = 224, 224


def extract_features(train_data_dir, nb_train_samples, batch_size, save_prefix):

    Itemcodes = []

    datagen = ImageDataGenerator(rescale=1. / 255)
    generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)

    for filename in generator.filenames:
        item_code = filename.split('/')[-1].split('.')[0]
        Itemcodes.append(item_code)

    # Modelo pré-treinado com pooling='avg' para todos os gêneros
    model = ResNet50(include_top=False, weights='imagenet', pooling='avg')

    extracted_features = model.predict(generator, steps=nb_train_samples // batch_size, verbose=1)
    print(f"Shape das características extraídas: {extracted_features.shape}")

    np.save(f'./{save_prefix}_ResNet_features.npy', extracted_features)
    np.save(f'./{save_prefix}_ResNet_feature_product_ids.npy', np.array(Itemcodes))
    print("Características extraídas e salvas com sucesso.")


# Boys
train_data_dir_boys = "/content/fashion-images/data/Apparel/Boys/Images"
nb_train_samples_boys = 759
extract_features(train_data_dir_boys, nb_train_samples_boys, 1, "Boys")

# Girls
train_data_dir_girls = "/content/fashion-images/data/Apparel/Girls/Images"
nb_train_samples_girls = 567
extract_features(train_data_dir_girls, nb_train_samples_girls, 1, "Girls")

# Men
train_data_dir_men = "/content/fashion-images/data/Footwear/Men/Images/"
nb_train_samples_men = 811
extract_features(train_data_dir_men, nb_train_samples_men, 1, "Men")

# Women
train_data_dir_women = "/content/fashion-images/data/Footwear/Women/Images/"
nb_train_samples_women = 769
extract_features(train_data_dir_women, nb_train_samples_women, 1, "Women")
```

```python
from sklearn.metrics.pairwise import pairwise_distances
from IPython.display import Image, display
import os

def get_similar_products_cnn(product_id, num_results):
    try:
        doc_id = list(all_Productids).index(product_id)
    except ValueError:
        print(f"Product ID '{product_id}' not found in the feature list.")
        return  # Exit if product ID not found

    pairwise_dist = pairwise_distances(all_extracted_features, all_extracted_features[doc_id].reshape(1, -1))
    indices = np.argsort(pairwise_dist.flatten())[0:num_results]
    pdists = np.sort(pairwise_dist.flatten())[0:num_results]

    print("="*20, "Input Product Image", "="*20)
    input_product = fashion_df.loc[fashion_df["ProductId"] == product_id]
    display(input_product)

    image_filename = f"{product_id}.jpg"
    image_path = None
    for root, dirs, files in os.walk("/content/fashion-images/data"):
        if image_filename in files:
            image_path = os.path.join(root, image_filename)
            break

    if image_path:
        display(Image(filename=image_path, width=224, height=224, embed=True))
    else:
        print(f"Image file '{image_filename}' not found.")

    print("\n", "="*20, "Recommended Products", "="*20)
    for i in range(1, len(indices)):
        recommended_product_id = all_Productids[indices[i]]
        recommended_product = fashion_df.loc[fashion_df["ProductId"] == recommended_product_id]

        # Check if recommended_product is empty before accessing it
        if not recommended_product.empty:
            image_filename = f"{recommended_product_id}.jpg"
            image_path = None
            for root, dirs, files in os.walk("/content/fashion-images/data"):
                if image_filename in files:
                    image_path = os.path.join(root, image_filename)
                    break

            if image_path:
                display(Image(filename=image_path, width=224, height=224, embed=True))
                print("Product Title:", recommended_product["ProductTitle"].iloc[0])
                print("Euclidean Distance from input image:", pdists[i])
            else:
                print(f"Image file '{image_filename}' not found.")
        else:
            print(f"Product ID '{recommended_product_id}' not found in fashion_df.")
```
